```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('nutrition.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
df
```

```python
df['Calories'] = df['Calories'].apply(lambda x: float(x.split(' ')[0]))
df['Protein'] = df['Protein'].apply(lambda x: float(x.split(' ')[0]))
df['Carbohydrate'] = df['Carbohydrate'].apply(lambda x: float(x.split(' ')[0]))
df['Total fat'] = df['Total fat'].apply(lambda x: float(x.split(' ')[0]))
df['Cholesterol'] = df['Cholesterol'].apply(lambda x: float(x.split(' ')[0]))
df['Fiber'] = df['Fiber'].apply(lambda x: float(x.split(' ')[0]))
df['Water'] = df['Water'].apply(lambda x: float(x.split(' ')[0]))
df['Alcohol'] = df['Alcohol'].apply(lambda x: float(x.split(' ')[0]))
df['Vitamin C'] = df['Vitamin C'].apply(lambda x: float(x.split(' ')[0]))
df.head()
```

```python
df.to_csv('nutrition_prepro1.csv', index = False)
```
